Exercicio 4.2 Escreva um programa que pergunte a velocidade do carro de um usuário. Caso ultrapasse 80km/h, exiba uma mensagem dizendo que o usuario foi multado. Nesse caso o valor da multa, cobrando R$ 5 por km acima de 80 km/h

```python
velocidade = int(input("Digite a velocidade do carro: "))

if velocidade > 80:
  multa = velocidade - 80
  print(f"Voce foi multado em R$ {multa * 5: 4.2f}")
```

> Calcule o impot de renda

*   0,00 a 1.000,00 - 0%
*   1.000,00 a 3.000,00 - 20%
*   3.000,00 pra cima - 35%

```python
salario = float(input("Digite o salario para calculo do imposto de renda: "))
base = salario
imposto = 0

if salario > 3000:
  imposto = imposto + ((base - 3000)* 0.35)
  base = 3000 #esse base so sera executada se atender a condicao de salario maior que 3000
if base > 1000:
  imposto = imposto + ((base - 1000)* 0.20)

print(f"Salario R$ {salario: 6.2f}. Imposto a paar: R$ {imposto: 6.2f}")
```

Exercicio 4.6 Escreva um programa que pergunte a distância que um passageiro deseja percorrer em km. Calcule o preço da passagem cobrando 0,50 por km para viagens de até 200km, e 0,45 para viagens mais longas.

```python
distanciaPercorrida = int(input("Informe a distancia total percorrida em KM: "))

#determinando qual o valor da passagem
if distanciaPercorrida <= 200:
  pPassagem = 0.50
else:
  pPassagem = 0.45

#encontrando o valor total a ser pago
valorTotal = distanciaPercorrida * pPassagem

print(f"O valor total da passagem foi de R${valorTotal:5.2f}")
```

## 4.3 Estrutura aninhadas

Os planos da empresa oferem preços diferentes de acorod com a quantidade de minutos usados por mês. Abaxo de 200 minutos, a empresa cobra R 0,20 por minuto. Entre 200 e 400 minutos, o preço é de R 0,18. Acima de 400 minutos, o preço por minuto é de R 0,15

```python
minutos = int(input("Quantos minutos você utilizou este mês:"))
if minutos < 200:
  preco = 0.20
else:
  if minutos < 400:
    preco = 0.18
  else:
    preco = 0.15

print(f"Você vai pagr este mês: R$ {minutos* preco: 6.2f}")
```

Exercício 4.8

Escreva um programa que leia dois números e que pergunte qual operação você deseja realizar. Você deve poder calcular soma (+), subtração (-), multiplicação (*) e divisão (/). Exiba o resultado da operação solicitada.

```python
primeiro_Numero = float(input("Digite o primeiro numero: "))
segundo_Numero = float(input("Digite o segundo numero: "))

tipoOperacao = int(input("Informe qual tipo de operação que você deseja realizar. \n 1. Soma \n 2. Subtração \n 3. Multiplicação \n 4. Divisão: \n"))

if tipoOperacao == 1:
  resultado = primeiro_Numero + segundo_Numero
elif tipoOperacao == 2:
  resultado = primeiro_Numero - segundo_Numero
elif tipoOperacao == 3:
  resultado = primeiro_Numero * segundo_Numero
elif tipoOperacao == 4:
  resultado = primeiro_Numero / segundo_Numero
else:
  print(f"Tipo de operação incorreta, selecione os valores entre 1 e 4")
  resultado = 0

print(f"O resultado da operação foi de {resultado}")
```

```python
# Código escritor

a = float(input("Primeiro número:"))
b = float(input("Segundo número:"))
operação = input("Digite a operação a realizar (+, -, * ou /):")
if operação == "+":
    resultado = a + b
elif operação == "-":
    resultado = a - b
elif operação == "*":
    resultado = a * b
elif operação == "/":
    resultado = a / b
else:
    print("Operação inválida!")
    resultado = 0
print("Resultado: ", resultado)
```

Exercício 4.9

Escreva um programa que aprovar o emprétimo bancário para comprar de uma casa. O programa deve perguntar o valor da casa a comprar, o salário e a quantidade de anos a pagar. O valor da prestação mensal não pode ser superior a 30% do salário. Calcule o valor da prestação como sendo o valor da casa a comprar dividido pelo número de meses a pagar.

```python
valor_imovel = float(input("Digite o valor do imóvel: R$ "))
remuneracao = float(input("Informe o valor da remuneracao atual: R$ "))
qtd_anos_financiamento = int(input("Informe a quantidade de anos que sera realizado financiamento: "))

valor_prestacao = valor_imovel / (qtd_anos_financiamento * 12)

if valor_prestacao > (remuneracao * 0.30):
  print(f"Financiamento recusado, pois o valor da parcela excedeu o limite da renda, valor excedente R$ {valor_prestacao - (remuneracao * 0.30): 4.2f}")
else:
  print(f"Financiamento aprovado, o valor das parcelas sera de R$ {valor_prestacao: 4.2f} durante o periodo de {qtd_anos_financiamento} anos.")

```
